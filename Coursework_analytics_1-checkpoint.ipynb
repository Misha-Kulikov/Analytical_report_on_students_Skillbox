```python
# Импортируем библиотеки, которые будут нужны для выполнения заданий.
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Загружаем в Ноутбук файлы с данными.
courses = pd.read_csv('courses.csv')
students = pd.read_csv('students.csv')
course_contents = pd.read_csv('course_contents.csv')
progresses = pd.read_csv('progresses.csv')
# В Датасете "progress_phases" есть колонки с датами. При загрузке сразу приведём их к формату дат.
progress_phases = pd.read_csv('progress_phases.csv', parse_dates=['start_date', 'finish_date'], dayfirst=False)
```

```python
# Общий Датасет

# Объединяем ДатаФреймы с "прогрессом"
union_progress = progress_phases.merge(progresses, left_on='progress_id', right_on='id')
# Убираем дублирующую колонку с id прогресса. Убираем "просто id", потому что далее ещё будут колонки с id.
progress_name_col = list([i for i in union_progress.columns if i != 'id'])
# Обновляем ДатаФрейм, убираем колонку "id"
union_progress = union_progress[progress_name_col]

# Объединяем ДатаФреймы с "курсами"
union_courses = course_contents.merge(courses, left_on='course_id', right_on='id')
# Убираем дублирующую колонку с id курса. Убираем "просто id", потому что далее ещё будут колонки с id.
courses_name_col = list([i for i in union_courses.columns if i != 'id'])
# Обновляем ДатаФрейм, убираем колонку "id"
union_courses = union_courses[courses_name_col]

# Объединяем "погресс" и "курсы" в единый ДатаФрейм
big_df = union_progress.merge(union_courses, how='inner') 
big_df.head()
```

```python
# Датасет со студентами

# Добавляем к объединенному ДатаФейму ДатаФрейм со студентами
students_df = big_df.merge(students, left_on='student_id', right_on='id')
# Убираем дублирующие столбцы
students_list = list([i for i in students_df.columns if i != 'id' and i != 'id_'])
# Обновляем ДатаФрейм
students_df = students_df[students_list]
# Меняем данные в "birthday" со str на DateTime. Т.к. есть несколько дат со странными годами, то заменяем их на пропущенные 
# значения
students_df['birthday'] = pd.to_datetime(students_df['birthday'], errors='coerce')
# Убираем строки с пропущенными данными
students_df.dropna(inplace=True)

# Определяем возраст студентов

# У нас достаточно старый Датасет (сейчас август 2022, а последний прогресс у студентов датируется 17.07.2019), 
# чтобы не прибавлять студентам "лишние" 3 года - будем расчитывать возраст не от актуальной сегодняшней даты, 
# а от самой поздней даты прогресса в данных. Брать её будем из столбца "finish_date".
my_today = big_df.finish_date.max()
age_list = []
for i in students_df['birthday']:
    age = my_today.year - i.year
    if my_today.month > i.month:
        age -= 1
    if my_today.month == i.month:
        if my_today.day > i.day:
            age -= 1
    age_list.append(age)

students_df['age'] = age_list # Добаовляем возраст к ДатаФрейму
```

```python
# Считаем значения из первого пункта задания.

course_count = big_df.title.drop_duplicates().count()
module_count = big_df[['title', 'module_title']].drop_duplicates().groupby('title').count().sort_values('module_title')
lesson_count = big_df[['title', 'module_title', 'lesson_token']].drop_duplicates().groupby(['title', 'module_title']).count().sort_values('lesson_token', ascending=False)
median_lesson_count = lesson_count.groupby('title').median().sort_values('lesson_token', ascending=False)
student_count = big_df[['title', 'student_id']].drop_duplicates().groupby('title').count().sort_values('student_id', ascending=False)
students_age = students_df['age'].agg(['min', 'max', 'mean', 'median'])
students_courses_age = students_df[['title', 'age']].groupby('title').agg(['min', 'max', 'mean', 'median'])
```

```python
# Создаем bar-chart из второго пункта задания:
# Постройте bar-chart, отражающий количество студентов на каждом курсе. Ticks нужно развернуть так, чтобы они были читаемы.

# Созадем Датасет с данными для графика.
student_count_bar1 = big_df[['title', 'student_id']].drop_duplicates().groupby('title').agg('count')\
                        .sort_values('student_id', ascending=False)
student_count_bar1.reset_index(inplace=True)

plt.bar(student_count_bar1['title'], student_count_bar1['student_id'])
plt.title('Количество студентов на курсе', pad = 10)
plt.ylabel('Количество студентов')
plt.xlabel('Название курса')
plt.xticks(rotation=90)
plt.show()
```

```python
# Создаем горизонтальный bar-chart из третьего путкта задания:
# Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий количество студентов на каждом 
# курсе. График должен иметь заголовок. Значения должны быть отсортированы. Цвет столбцов должен содержать информацию о сфере,
# к которой относится курс (то есть нужна легенда). Прозрачность должна стоять на отметке 0.1. На график должна быть нанесена 
# линия медианы. У медианы должен быть свой цвет. Рамки у графика быть не должно ⭐

# Созадем Датасет с данными для графика.
student_count_bar2 = big_df[['title', 'student_id', 'field']].drop_duplicates().groupby(['title', 'field']).agg('count')\
                        .sort_values('student_id', ascending=False)
student_count_bar2.reset_index(inplace=True)

# В задании указано, что прозрачность столбцов у гистограммы должна быть 10%. Но если выполнять это условие, то столбцы будут
# очень слабо видны. Поэтому, чтобы улучшить читаемость графика, прозрачность столбцов гистограммы установил на уровне 60%.
# Это минимальный уровень прозрачности, на котором данные с графика можно комфортно считать.
sns.barplot(data=student_count_bar2, y='title', x='student_id', hue='field', orient='h', dodge=False, alpha = 0.6)
plt.title('Количество студентов на курсе', pad = 10)
plt.ylabel('Название курса')
plt.xlabel('Количество студентов')
plt.axvline(x = student_count_bar2.student_id.median(), color = 'red')
sns.despine(top=True, right=True) # Убираем верхнию и правую границы, осталяем только оси
plt.legend()
plt.show()
```

```python
# Подготавливаем Датасет для графиков с приростом студентов на курсах.

# Создаем отдельный Датафрейм и помещаем в него только необходимые данные:
# - Отфильровываем по студентам, которые начали хотя бы одну домашнююю работу;
# - Берем только нужные столбцы - уникальные ID студентов, даты старта и название курса.
# Сортируем их по дате, чтобы потом удалить дубликаты.
student_growth = big_df[big_df['is_homework']==True][['student_id', 'start_date', 'title']].sort_values(['student_id', 'title', 'start_date'])

# Оставляем только уникальные значения студентов в каждом курсе через удаление дубликатов.
student_growth.drop_duplicates(subset=['student_id', 'title'], inplace=True)

# Округляем даты до месяц/год через единое число в каждом месяце.
student_growth['start_date'] = [i.replace(day=1, hour=0, minute=0, second=0, microsecond=0, nanosecond=0) for i in student_growth['start_date']]

# Считаем количество студентов на каждом курсе в каждом месяце через группировку по признаку и агрегирование.
student_growth = student_growth.groupby(['start_date', 'title']).agg('count')

# Возвращщаем значения из Мультииндекса в обычные стобцы Датафрейма, чтобы можно было указать оси для графиков.
student_growth.reset_index(inplace=True)
```

```python
# Подготавливаем Датасет для графиков с прогрессом по курсам и месяцам.

# Создаем отдельный Датафрейм и помещаем в него только необходимые данные.
progress = big_df[big_df['is_homework']==True][['start_date', 'finish_date', 'title']]

# Округляем даты до месяц/год через единое число в каждом месяце.
progress['start_date'] = [i.replace(day=1, hour=0, minute=0, second=0, microsecond=0, nanosecond=0) for i \
                          in progress['start_date']]
progress['finish_date'] = [i.replace(day=1, hour=0, minute=0, second=0, microsecond=0, nanosecond=0) for i \
                           in progress['finish_date']]

# Удаляем работы, которые не были завершены (в столбце 'finish_date' есть значения NaN).
progress.dropna(inplace=True)

# Переопределяем нумерацию в индексе, чтобы удобнее было работать дальше в цикле.
progress.reset_index(drop=True, inplace=True)

# Считаем количество прогрессов без учета переходящих месяцев, для каждого курса и месяца.
count_progress = progress.groupby(['title', 'start_date']).count()

# Пробегаемся циклом по всем данным, чтобы добавить данные по переходящим месяцам.
for i in progress.index:
    if progress.iloc[i]['start_date'] != progress.iloc[i]['finish_date']: # Работаем только с переходящими месяцами
        start_month = progress.iloc[i]['start_date'].month # Определяем месяц старта прогресса
        finish_month = progress.iloc[i]['finish_date'].month # Определяем месяц финиша прогресса
        while start_month < finish_month: # Добовляем по 1 прогрессу к каждому месяцу между стартом и финишем
            start_month += 1 # Месяц старата уже посчитан, поэтому прогресс добавляем к следующему месяцу
            date = progress.iloc[i]['start_date'].replace(month=start_month) # Определяем к какому месяцу добовляем 1 прогресс
            course = progress.iloc[i]['title'] # Определяем к какому курсу добовляем 1 прогресс
            try:
                count_progress.loc[(course, date)] += 1 # добовляем +1 прогресс к имеющимся данным
            except:
                count_progress.at[(course, date), 'finish_date'] = 1 # если данные отсутствуют, дополняем данные новой строкой

# Возвращщаем значения из Мультииндекса в обычные стобцы Датафрейма, чтобы можно было указать оси для графиков.
count_progress.reset_index(inplace=True)

# Сортируем значения по дате старта, чтобы на графиках значения отображались в порядке возрастания дат.
count_progress.sort_values(['title', 'start_date'], inplace=True)
```

## 4. Расчет конверсии

1. **(p)** Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).
2. **(m)** Постройте bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. График должен иметь заголовок. Ticks нужно развернуть так, чтобы они были читаемы
3. **(m)** Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. 15 графиков. Графики должны иметь заголовки. Ticks должны содержать номер и название модуля. Цвет столбцов графиков должен содержать информацию о сфере, к которой относится курс (нужна легенда). Прозрачность должна стоять на отметке 0.1. На графики должна быть нанесена линия медианы конверсии для каждого курса. У медианы должен быть свой цвет. Рамок у графиков быть не должно ⭐
4. На основании рассчитанных значений сделайте аналитический вывод (должен быть полным и покрывать все полученные выше метрики)

```python
# Подготавливаем Датасет для графиков с конверсией.

# Создаем отдельный Датасет с данными, которые нам потребуются для расчета конверсии.
# Т.к. в 3-ем задании на графике цветом нужно будет отразить сферу, к которой относится курс, то дополнительно берём ещё и 
# столбец "field".
# Переопределяем индекс для удобной работы в цикле.
conversion = big_df[big_df['is_homework'] == True][['field', 'title', 'module_number', 'module_title', \
                                                    'lesson_number', 'lesson_title', 'student_id']]
conversion.reset_index(inplace=True)

# Создаем отдельный Датасет, в котором будет информация только по номерам уроков в каждом модуле каждого курса, 
# чтобы в цикле быстро находить первый и последний номер урока с ДЗ.
all_lessens = conversion[['title', 'module_number', 'module_title', 'lesson_number', 'lesson_title']]
all_lessens.drop_duplicates(subset=['title', 'module_number', 'module_title', 'lesson_number', 'lesson_title'], \
                            inplace=True, ignore_index=True)

# Определяем является ли урок первым в модуле и добавляем эту информацию в Датасет для расчета конверсии.
list_start = [] # Заготовка
for i in conversion.index: # Проходимся циклом по каждой записи Датасета с конверсией
    title = conversion.iloc[i].title # Определяем название курса у записи
    module_number = conversion.iloc[i].module_number # Определяем номер модуля у записи
    lesson_i = conversion.iloc[i].lesson_number # Определяем номер урока у записи
    # Находим номер первого урока для курса и модуля из записи
    lesson_n = all_lessens[(all_lessens.title == title) & (all_lessens.module_number == module_number)].lesson_number.min()
    if lesson_i == lesson_n: # Если номер урока записи равен номеру первого урока, то это первый урок
        list_start.append(True)
    else: # Если не равен, то это не первый урок
        list_start.append(False)
conversion['lesson_start'] = list_start # Добавляем полученый список в Датафрейму с конверсией

# Определяем является ли урок последним в модуле и добавляем эту информацию в Датасет для расчета конверсии.
list_finish = [] # Заготовка
for i in conversion.index: # Проходимся циклом по каждой записи Датасета с конверсией
    title = conversion.iloc[i].title # Определяем название курса у записи
    module_number = conversion.iloc[i].module_number # Определяем номер модуля у записи
    lesson_i = conversion.iloc[i].lesson_number # Определяем номер урока у записи
    # Находим номер последнего урока для курса и модуля из записи
    lesson_n = all_lessens[(all_lessens.title == title) & (all_lessens.module_number == module_number)].lesson_number.max()
    if lesson_i == lesson_n: # Если номер урока записи равен номеру последнего урока, то это последний урок
        list_finish.append(True)
    else: # Если не равен, то это не последний урок
        list_finish.append(False)
conversion['lesson_finish'] = list_finish # Добавляем полученый список к Датафрейму с конверсией
```

```python
# Создаем отдельный Датафрейм для подсчета количества студетнтов на первых уроках каждого модуля каждого курса.
conversion_start = conversion[conversion['lesson_start'] == True]\
                            [['field', 'title', 'module_number', 'module_title', 'student_id']]\
                            .groupby(['field', 'title', 'module_number', 'module_title']).agg('count')
conversion_start.sort_values(['field', 'title', 'module_number'], inplace=True)
conversion_start.reset_index(inplace=True)

# Меняем название у столбца, чтобы более точно отобразить смысл данных в этом столбце.
conversion_start.rename(columns={'student_id': 'students_count'}, inplace=True)

# У курса 'UX-дизайн' в 9-ом модуле есть 2 домашние работы.
# Вторую домашнюю работу добавили уже после того, как некоторые студенты успели закончить следующий модуль и даже весь курс.
# Количество студентов, которые сдали вторую домашнюю работу в 9 модуле меньше, чем количество студентов, которые сдали первую 
# домашнюю работу в следующем модуле.
# У нас получится, что конверсия из 9 модуля в 10 модуль будет выше 100%, что не корректно.
# Чтобы этого избежать мы будем считать, что в 9 модуле была только 1 домашняя рабта, которую сдавали все студенты, 
# вне зависимости от даты прохождения курса.
# Для этого пометим, что 2 домашняя работа (это урок №8 в модуле) не является последней домашней работой в модуле.
conversion['lesson_finish'][(conversion['title'] == 'UX-дизайн') & (conversion['module_number'] == 9) \
                            & (conversion['lesson_number'] == 8)] = False
# Отметим, что первая домашняя работа (это урок №7 в модуле) является последней домашней работой.
conversion['lesson_finish'][(conversion['title'] == 'UX-дизайн') & (conversion['module_number'] == 9) \
                            & (conversion['lesson_number'] == 7)] = True

# Создаем отдельный Датафрейм для подсчета количества студентов на последних уроках каждого модуля каждого курса.
conversion_finish = conversion[conversion['lesson_finish'] == True]\
                                        [['field', 'title', 'module_number', 'module_title', 'student_id']]\
                                        .groupby(['field', 'title', 'module_number', 'module_title']).agg('count')
conversion_finish.sort_values(['field', 'title', 'module_number'], inplace=True)
conversion_finish.reset_index(inplace=True)

# Меняем название у столбца, чтобы более точно отобразить смысл данных в этом столбце.
conversion_finish.rename(columns={'student_id': 'students_count'}, inplace=True)
```

```python
# Добавляем количество студентов, которые завершили домашние работы в стартовый ДатаФрейм, в котором будем считать конверсию.
lesson_finish = []
for i in conversion_start.index:
    title = conversion_start.iloc[i].title # Определяем название курса у записи
    module_number = conversion_start.iloc[i].module_number # Определяем номер модуля у записи
    # .
    students_finish = conversion_finish[(conversion_finish['title'] == title) \
                                        & (conversion_finish['module_number'] == module_number)].students_count.iloc[0]
    lesson_finish.append(students_finish) # Добавляем номер в список
conversion_start['students_finish'] = lesson_finish # Добавляем полученый список к Датафрейму с конверсией

# Меняем название у столбца, чтобы более точно отобразить смысл данных в этом столбце.
conversion_start.rename(columns={'students_count': 'students_start'}, inplace=True)

# Отсортируем список по убыванию количества студентов из модуля в модуль для каждого курса.
conversion_start.sort_values(['title', 'students_start'], inplace=True, ascending=False, ignore_index=True)

# Считаем конверсию.
# Создаем столбец для расчета конверсии, по умолчанию считаем что конверсия 100%.
conversion_start['students_conversion'] = 1
for i in conversion_start.index[1:]: # Проходим циклом по каждой записи.
    # Для первых модулей каждого курса конверсия 100%.
    # Если это не первый модуль в курсе, то расчитываем конверсию.
    if conversion_start.title.iloc[i] == conversion_start.title.iloc[i-1]:
        conversion_start.students_conversion.iloc[i] = conversion_start.students_start.iloc[i] \
                                                        / conversion_start.students_finish.iloc[i-1]

# Добовляем новый столбец с объединенными номером и названием модуля для второго задания с графиками.
# Сначала приводим численные значения в строковые, чтобы можно было соединять строки со строками.
conversion_start['module_number_title'] = conversion_start['module_number'].apply(str)
# Потом добовляем название модудя к номеру модуля.
conversion_start['module_number_title'] = conversion_start['module_number_title'] + ' ' + conversion_start['module_title']
```

### Первая метрика

```python
# Сначала разберемся со студентами, которые не закончили какие-то уроки.

# Судя по заданию с конверсией, у нас большое количество студентов, которые остановились на каком-то модуле и не закончили 
# обучение на курсе. Не законченный курс, модуль или домашнее задание само по себе не означает, что студент плохо справляется с 
# программой обучения, возможно, он завершит задание в ближайшем будущем и уложится в обычные сроки выполнения (средние сроки 
# выполнения заданий у других студентов).

# Определим студентов которые не завершили урок в отдельный ДатаФрейм.
no_finish_students = big_df[big_df['finish_date'].isnull() == True]
no_finish_students.reset_index(inplace=True, drop=True)

# У нас достаточно старый Датасет (сейчас август 2022, а последний прогресс у студентов датируется 17.07.2019), поэтому 
# определим самую познюю дату програсса и будем использовать её как крайнюю точку для расчета количества дней, которые студент 
# выполняет работу (сколько он уже не сдает работу, в днях, аналог "finish_date").
finish_date_max = big_df.finish_date.max()
no_finish_students['no_finish_duration'] = finish_date_max - no_finish_students['start_date']
# Переводим данные из типа Timedelta в int для большего удобства работы.
no_finish_students['no_finish_duration'] = [i.days for i in no_finish_students['no_finish_duration']]

# Чтобы выполнить домашнее задание или посмотреть урок нужно время. Определим какое количество времени является нормой для 
# выполнения задания, а какое уже превышение нормы. 
# Создаем новый столбец со временем прохождения каждого урока у каждого студента.
big_df['lesson_duration'] = big_df['finish_date'] - big_df['start_date']
# Округляем время прохождения до дней.
big_df['lesson_duration'] = [i.days for i in big_df['lesson_duration']]
# За норму возьмем 90 квантиль длительности каждого урока, рассмотрим самых "заядлых прогульщиков".
df_lesson_dur_q90 = big_df[['title', 'module_number', 'lesson_number', 'lesson_duration']]\
                            .groupby(['title', 'module_number', 'lesson_number']).quantile(q=0.90)

# Добавим значение "нормы" (90 квантиль) к нашему ДатаФрейму со студентами, которые не завершили начатый урок.
lesson_dur_norm_list = [] # Заготовка
for i in no_finish_students.index: # Проходимся циклом по каждой записи ДатаФрейма
    title = no_finish_students.iloc[i].title  # Определяем название курса у записи
    module_number = no_finish_students.iloc[i].module_number # Определяем номер модуля у записи
    lesson_number = no_finish_students.iloc[i].lesson_number # Определяем номер урока у записи
    lesson_dur_norm = df_lesson_dur_q90.loc[(title, module_number, lesson_number)][0] # Находим значение нормы для записи
    lesson_dur_norm_list.append(lesson_dur_norm)  # Добавляем значение нормы в нашу заготовку
no_finish_students['lesson_dur_norm'] = lesson_dur_norm_list # Добавляем полученый список к Датафрейму 

# Расчитываем на сколько дней выполнение урока у студента отклоняется от нашей нормы.
no_finish_students['deviation_norm'] = no_finish_students['no_finish_duration'] - no_finish_students['lesson_dur_norm']

# Подготавливаем список студентов, которые тратят больше времени на выполнение урока, чем наша норма.
# Для большей информативности добавим данные о том, на каких курсах/модуля/уроках остановился наш проблемный студент, когда 
# (дата) студент остановился, сколько дней он выполняет урок, на котором остановился, какова норма выполнения этого урока и
# насколько (в днях) оклонился от нормы.
problem_students = no_finish_students[no_finish_students['deviation_norm'] > 0]\
[['student_id', 'title', 'module_number', 'module_title', 'lesson_number', 'lesson_title', 'is_homework', 'start_date', \
  'no_finish_duration', 'lesson_dur_norm', 'deviation_norm']]\
.groupby(['student_id', 'title', 'module_number', 'module_title','lesson_number', 'lesson_title', 'is_homework'])\
.agg({'start_date': 'max', 'no_finish_duration': 'sum', 'lesson_dur_norm': 'sum', 'deviation_norm': 'sum'})

problem_students.head()
```

### Вторая метрика

```python
# Чтобы выполнить домашнее задание или посмотреть урок нужно время. Определим какое количество времени является нормой для 
# выполнения задания, а какое уже превышение этой нормы. 
# За норму возьмем 75 квантиль длительности каждого урока. Если брать медиану, то мы запишем в проблемные студенты тех, кто  
# совсем немного отклонился от нормы, что не совсем верно. Если взять 90 квантиль, то мы сильно занизим норму по завершению 
# урока. Поэтому берем среднее - 75 квантиль.
# Расчитываем норму выполнения урока для каждого урока в каждом модуле каждого курса.
df_lesson_dur_q75 = big_df[['title', 'module_number', 'lesson_number', 'lesson_duration']]\
                            .groupby(['title', 'module_number', 'lesson_number']).quantile(q=0.75)
# Для работы в цикле возвращаем столбцы из Мультииндекса в обычные столбцы. 
df_lesson_dur_q75.reset_index(inplace=True)

# Теперь нам нужно добавить значения нормы в основной ДатаФрейм. 
# В основном ДатаФрейме ('big_df') более 350 тыс. строк, а в Датасете с нормой 1200 строк. Код быстрее сработает, если мы будем 
# добавлять значение нормы для групп студентов, нежели для каждого студента отдельно. Поэтому циклом проходим по Датасету с 
# нормой.

# Создаем столбец в основном ДатаФрейме, определяем, что значения в нём будут типа float.
big_df.loc[:, 'lesson_dur_norm'] = 0.0

# Добавим значение "нормы" (75 квантиль) к нашему ДатаФрейму со студентами, которые не завершили начатый урок.
for i in df_lesson_dur_q75.index: # Проходимся циклом по каждой записи ДатаФрейма
    title = df_lesson_dur_q75.iloc[i].title # Определяем название курса у записи
    module_number = df_lesson_dur_q75.iloc[i].module_number # Определяем номер модуля у записи
    lesson_number = df_lesson_dur_q75.iloc[i].lesson_number # Определяем номер урока у записи
    lesson_dur_norm = df_lesson_dur_q75.iloc[i].lesson_duration # Находим значение нормы для записи
    # Добавляем значение нормы в основной ДатаФрейм
    big_df.loc[(big_df['title'] == title) & (big_df['module_number'] == module_number) & \
               (big_df['lesson_number'] == lesson_number), 'lesson_dur_norm'] = lesson_dur_norm

# Расчитываем отклонение от нормы для кажого прогресса каждого студента.
big_df['deviation_norm'] = big_df['lesson_duration'] - big_df['lesson_dur_norm']

# Создаем Дотасет с проблемными студентами.
# Если брать метрику "больше 3-ёх отклонений от нормы у студента по всем урокам на курсе = прблемный студент", то получится 
# больше 6 тыс. такиех студентов (68% от общего числа студентов). Это очень много и разбираться в проблемах сразу всех этих 
# студентов слишком долго и проблемно. Поэтому сначала будем рассматривать студентов которые имеют больше 3 отклонений в сдаче
# домашней работы.

# Определяем для каждого студента количество превышений нормы в сдаче домашних работ.
problem_students_metrics = big_df.loc[(big_df['is_homework'] == True) & (big_df['deviation_norm'] > 0), \
                                      ['student_id', 'title', 'deviation_norm']].groupby(['student_id', 'title']).agg('count')
# Оставляем студентов, которые больше 3-ёх раз отклонились от нормы.
problem_students_metrics = problem_students_metrics[problem_students_metrics['deviation_norm'] > 3]
# Получился Датасет со списком прооблемных студентов и курсов, на которых они столкнулись с трудностями в выполении домашних 
# работ.

# Создаем список студентов, для этого удаляем дубликаты в стобце 'student_id'.
problem_students_metrics_list = problem_students_metrics.reset_index().student_id.drop_duplicates()
problem_students_metrics_list
# Получилось 709 проблемных студентов (7% от общего числа студентов).
```
